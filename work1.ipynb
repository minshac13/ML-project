```python
import pandas as pd
```

```python
data = pd.read_csv('cardio_data.csv')
data
```

```python
data.drop('id',inplace=True,axis=1)
data.drop('age',inplace=True,axis=1)
```

```python
data = data.rename(columns={
    'ap_hi': 'systolic_bp', 
    'ap_lo': 'diastolic_bp'
})
```

```python
data.drop('bp_category_encoded',axis=1,inplace=True)
```

```python
data.info()
```

```python
data['bp_category']=data['bp_category'].replace({'Normal': 0,
    'Elevated': 1,
    'Hypertension Stage 1': 2,
    'Hypertension Stage 2': 3}).astype(float)
```

```python
data.info()
```

```python
data['gender']=data['gender'].ffill()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(data[['gender']])
gender = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
print(gender)
```

```python
data.drop('gender',axis=1,inplace=True)

data = pd.concat([data,gender],axis=1)
data.head()
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select columns that are actually numeric (excluding ID and Target)
cols_to_plot = ['age_years', 'height', 'weight', 'systolic_bp', 'diastolic_bp', 'bmi']

# Set up the figure and grid
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.suptitle('Outlier Detection for Cardiovascular Features', fontsize=16)

for i, col in enumerate(cols_to_plot):
    row = i // 3
    col_idx = i % 3
    sns.boxplot(ax=axes[row, col_idx], y=data[col], palette="Set2")
    axes[row, col_idx].set_title(f'Distribution of {col}')
    axes[row, col_idx].set_ylabel('')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
cols = ['age_years', 'height', 'weight', 'systolic_bp', 'diastolic_bp', 'bmi']

for col in cols:
    q1 = data[col].quantile(.25)
    q3 = data[col].quantile(.75)
    iqr = q3 - q1
    lw = q1 - (iqr * 1.5)
    uw = q3 + (iqr * 1.5)
    
    outliers_count = len(data[(data[col] > uw) | (data[col] < lw)])
    print(f"{col}: {outliers_count} outliers (Bounds: {lw:.1f} to {uw:.1f})")
```

```python
# Remove impossible heights (Adults shorter than 120cm or taller than 250cm)
data = data[(data['height'] >= 120) & (data['height'] <= 250)]

# 2. Remove impossible weights (Adults lighter than 30kg)
data = data[data['weight'] >= 30]

# Fix BMI logic
# A BMI over 80 or under 10 is usually a measurement error.
data = data[(data['bmi'] >= 10) & (data['bmi'] <= 80)]

# Blood Pressure Logic Check
# Diastolic (low) should NEVER be higher than Systolic (high). 
# If it is, the data entry was swapped.
data = data[data['systolic_bp'] > data['diastolic_bp']]
```

```python
data.shape
```

```python
x = data.drop('cardio',axis=1)
y = data['cardio']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

print(f"Training shapes: {x_train.shape}, {y_train.shape}")
print(f"Testing shapes: {x_test.shape}, {y_test.shape}")
```

```python
x_train
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# 1. Initialize the Imputer
imputer = IterativeImputer(random_state=42)

# 2. Fit on X_train and transform it
x_train_imputed = imputer.fit_transform(x_train)
x_train_df = pd.DataFrame(x_train_imputed,columns=x_train.columns)

# 3. Transform X_test using the patterns learned from X_train
x_test_imputed = imputer.transform(x_test)
x_test_df = pd.DataFrame(x_test_imputed,columns=x_test.columns)
```

```python
x_train_df.isna().sum()
```

```python
x_test_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = pd.concat([x_train_df,y_train.reset_index(drop=True)],axis=1).corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Feature correlation Heatmap')
plt.show()
```

```python
x_train_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 3. Initialize the model
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

# 4. Train only on the 5 selected features
rf_model.fit(x_train_df, y_train)

# 5. Predict and Evaluate
y_pred = rf_model.predict(x_test_df)

print(f"Accuracy with every features: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
train_pred = rf_model.predict(x_train_df)
```

```python
from sklearn.metrics import accuracy_score

# 1. Calculate the scores
train_accuracy = accuracy_score(y_train, train_pred)
test_accuracy = accuracy_score(y_test, y_pred)

# 2. Calculate the difference (The Gap)
gap = train_accuracy - test_accuracy

print(f"--- Model Fit Analysis ---")
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy:  {test_accuracy:.4f}")
print(f"Accuracy Gap:      {gap:.4f}")

# 3. Simple Diagnostic Logic
if gap > 0.05:
    print("\nResult: Potential OVERFITTING. The model is memorizing training data.")
elif train_accuracy < 0.70 and test_accuracy < 0.70:
    print("\nResult: Potential UNDERFITTING. The model is too simple for this data.")
else:
    print("\nResult: GOOD FIT. The model generalizes well to new data.")
```
